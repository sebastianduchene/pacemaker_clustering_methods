```python
fig, axs = plt.subplots(1, 5)

svd_angio = np.linalg.svd(angio_dat)
cols = np.array(['r', 'b', 'y', 'g'])[list(angio_vbgmm1[1].predict(angio_dat))]
for i in range(angio_dat.shape[0]):
    axs[0].plot(svd_angio[0][i, 0], svd_angio[0][i, 1], cols[i]+'o', ms = 10, alpha =0.5)
    axs[0].set_title('Angiosperms')
    axs[0].set_ylabel('PC2')
    
svd_poac = np.linalg.svd(poac_data)
cols = np.array(['ro', 'bo', 'yo', 'go', 'r^', 'b^', 'y^', 'g^', 'r*', 'b*', 'y*', 'g*', 'r+', 'b+', 'y+', 'g+'])[list(poac_VBGMM[1].predict(poac_data))]
markers = np.array(['o', 's'])[list(poac_VBGMM[1].predict(poac_data))]
for i in range(poac_data.shape[0]):
    axs[1].plot(svd_poac[0][i, 0], svd_poac[0][i, 1], cols[i], ms = 10, alpha =0.5, marker = markers[i])
    axs[1].set_title('Poaceae')

eudicot_svd = np.linalg.svd(eudicot_dat)
cols = np.array(['r', 'b', 'g', 'm', 'c', 'k'])[np.array( eudicot_VBGMM[1].predict(eudicot_dat) )]
markers = np.array(['o', 's'])[np.array( eudicot_VBGMM[1].predict(eudicot_dat) )]
for i in range(eudicot_dat.shape[0]):
    axs[2].plot(eudicot_svd[0][i, 0], eudicot_svd[0][i, 1], cols[i]+'o', ms = 10, marker = markers[i], alpha =0.7)
    axs[2].set_title('Eudicots')
    axs[2].set_xlabel('PC1')

aster_svd = np.linalg.svd(aster_data)
cols = np.array(['ro', 'go', 'bo', 'yo', 'co', 'mo', 'ko', 'r^', 'y^', 'm^'])[list(aster_VBGMM[1].predict(aster_data))]
markers = np.array(['o', 'v', 's'])[list(aster_VBGMM[1].predict(aster_data))]
for i in range(aster_data.shape[0]):
    axs[3].plot(aster_svd[0][i, 0], aster_svd[0][i, 1], cols[i], ms = 10, alpha = 0.7, marker = markers[i])
    axs[3].set_title('Asteraceae')

rosids_svd = np.linalg.svd(rosids_data)
cols = np.array(['ro', 'bo', 'yo', 'go', 'mo', 'co', 'ko', 'k^'])[rosids_VGMM[1].predict(rosids_data)]
markers = np.array(['o', 's'])[rosids_VGMM[1].predict(rosids_data)]
for i in range(rosids_data.shape[0]):
    axs[4].plot(rosids_svd[0][i, 0], rosids_svd[0][i, 1], cols[i], ms = 10, alpha = 0.7, marker = markers[i])
    axs[4].set_title('Rosids')

pylab.savefig('figure_1.pdf')
```

```python
angio_genes = pd.read_csv('angiosperm_matrix.csv').index
poaceae_genes = pd.read_csv('poaceae_matrix.csv').index
eudicot_genes = pd.read_csv('eudicot_matrix.csv').index
astera_genes = pd.read_csv('asteraceae_matrix.csv').index
rosids_genes = pd.read_csv('rosids_matrix.csv').index

angio_clusters = pd.DataFrame(np.vstack([pd.read_csv('angiosperm_matrix.csv').index, angio_vbgmm1[1].predict(angio_dat)])).transpose()
poaceae_clusters = pd.DataFrame(np.vstack([pd.read_csv('poaceae_matrix.csv').index, poac_VBGMM[1].predict(poac_data)])).transpose()
eudicot_clusters = pd.DataFrame(np.vstack([pd.read_csv('eudicot_matrix.csv').index, eudicot_VBGMM[1].predict(eudicot_dat)])).transpose()
astera_clusters = pd.DataFrame(np.vstack([pd.read_csv('asteraceae_matrix.csv').index, aster_VBGMM[1].predict(aster_data)])).transpose()
rosids_clusters = pd.DataFrame(np.vstack([pd.read_csv('rosids_matrix.csv').index, rosids_VGMM[1].predict(rosids_data)])).transpose()

print [i.shape for i in [angio_genes, poaceae_genes, eudicot_genes, astera_genes, rosids_genes]]

def concat_dfs(df_list):
    import pandas as pd
    if len(df_list) == 2:
        return pd.merge(df_list[0], df_list[1], left_on = 0, right_on =0)
    elif len(df_list) > 2:
        return pd.merge(df_list[0], concat_dfs(df_list[1:]), left_on = 0, right_on =0)

merged_clusters = concat_dfs([angio_clusters, poaceae_clusters, eudicot_clusters, astera_clusters, rosids_clusters])
print merged_clusters.head()
```
