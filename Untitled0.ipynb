```python
## Angio data 
angio_dat = pd.DataFrame(rescale_data(np.array(pd.DataFrame(np.array(pd.read_csv('angiosperm_matrix.csv'))))))
angio_vbgmm1 = fit_VBGMM(np.array(angio_dat))
angio_dppgmm1 = fit_DPGMM(np.array(angio_dat))
print angio_vbgmm1[2]
print angio_dppgmm1[2]
```

```python
## Poaceae data
poac_data = pd.DataFrame(rescale_data(np.array(pd.read_csv('poaceae_matrix.csv'))))
poac_VBGMM = fit_VBGMM(np.array(poac_data))
poac_DPGMM = fit_DPGMM(np.array(poac_data))

print poac_VBGMM[2]
print poac_DPGMM[2]
```

```python
## Asteraceae data
aster_data = pd.DataFrame(rescale_data(np.array(pd.read_csv('asteraceae_matrix.csv'))))
aster_VBGMM = fit_VBGMM(np.array(aster_data))
aster_DPGMM = fit_DPGMM(np.array(aster_data))
print aster_VBGMM[2]
print aster_DPGMM[2] 
```

```python
## Eudicot data
eudicot_dat = pd.DataFrame(rescale_data(np.array(pd.read_csv('eudicot_matrix.csv'))))
eudicot_VBGMM = fit_VBGMM(np.array(eudicot_dat))
eudicot_DPP = fit_DPGMM(np.array(eudicot_dat))
print eudicot_VBGMM[2]
print eudicot_DPP[2]
```

```python
## Rosids data
rosids_data = pd.DataFrame(rescale_data(np.array(pd.read_csv('rosids_matrix.csv'))))
rosids_VGMM = fit_VBGMM(np.array(rosids_data))
rosids_DPP = fit_DPGMM(np.array(rosids_data))
print rosids_VGMM[2]
print rosids_DPP[2]
```

```python
fig, axs = plt.subplots(5, 2)

svd_angio = np.linalg.svd(angio_dat)
cols = np.array(['r', 'b', 'y', 'g'])[list(angio_vbgmm1[1].predict(angio_dat))]
for i in range(angio_dat.shape[0]):
    axs[0, 0].plot(angio_dat.ix[i, :], color = cols[i], alpha = 0.7, linewidth = 2)
    axs[0, 1].plot(svd_angio[0][i, 0], svd_angio[0][i, 1], cols[i]+'o', ms = 10, alpha =0.5)
    axs[0, 0].set_title('Angiosperms')

svd_poac = np.linalg.svd(poac_data)
cols = np.array(['ro', 'bo', 'yo', 'go', 'r^', 'b^', 'y^', 'g^', 'r*', 'b*', 'y*', 'g*', 'r+', 'b+', 'y+', 'g+'])[list(poac_VBGMM[1].predict(poac_data))]
for i in range(poac_data.shape[0]):
    axs[1, 0].plot(poac_data.ix[i, :], color = cols[i][0], alpha = 0.7, linewidth = 2)
    axs[1, 1].plot(svd_poac[0][i, 0], svd_poac[0][i, 1], cols[i], ms = 10, alpha =0.5)
    axs[1, 0].set_title('Poaceae')

aster_svd = np.linalg.svd(aster_data)
cols = np.array(['ro', 'yo', 'go', 'bo', 'co', 'mo', 'ko', 'r^', 'y^', 'm^'])[list(aster_VBGMM[1].predict(aster_data))]
for i in range(aster_data.shape[0]):
    axs[2, 0].plot(aster_data.ix[i, :], color = cols[i][0], linewidth = 2, alpha = 0.7)
    axs[2, 1].plot(aster_svd[0][i, 0], aster_svd[0][i, 1], cols[i], ms = 10, alpha = 0.7)
    axs[2, 0].set_ylabel('log10 branch length')
    axs[2, 1].set_ylabel('PC 1')
    axs[2, 0].set_title('Asteraceae')
    
eudicot_svd = np.linalg.svd(eudicot_dat)
cols = np.array(['r', 'b', 'g', 'm', 'c', 'k'])[np.array( eudicot_VBGMM[1].predict(eudicot_dat) )]
for i in range(eudicot_dat.shape[0]):
    axs[3, 0].plot(eudicot_dat.ix[i, :], color = cols[i], linewidth = 3, alpha = 0.7)
    axs[3, 1].plot(eudicot_svd[0][i, 0], eudicot_svd[0][i, 1], cols[i]+'o', ms = 10, alpha =0.7)
    axs[3, 0].set_title('Eudicots')
    
rosids_svd = np.linalg.svd(rosids_data)
cols = np.array(['ro', 'bo', 'yo', 'go', 'mo', 'co', 'ko', 'k^'])[rosids_VGMM[1].predict(rosids_data)]
for i in range(rosids_data.shape[0]):
    axs[4, 0].plot(rosids_data.ix[i, :], color = cols[i][0], linewidth = 3, alpha = 0.7)
    axs[4, 1].plot(rosids_svd[0][i, 0], rosids_svd[0][i, 1], cols[i], ms = 10, alpha = 0.7)
    axs[4, 0].set_xlabel('branch')
    axs[4, 1].set_xlabel('PC 2')
    axs[4, 0].set_title('Rosids')

    

pylab.savefig('figure_1.pdf')
```
