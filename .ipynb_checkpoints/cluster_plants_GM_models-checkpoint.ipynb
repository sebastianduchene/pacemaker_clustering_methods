```python
import os, sys, re
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import pylab
from pylab import rcParams
from sklearn import mixture
import scipy.stats as sp
from mpl_toolkits.mplot3d import Axes3D
from scipy.spatial import distance
import scipy.stats as sc
from GMM_trees import *
import itertools
%pylab inline
pylab.rcParams['figure.figsize'] = (20, 5)
```

## Angiosperms

```python
# Angiosperm 
angio_dat = pd.DataFrame(rescale_data(np.array(pd.DataFrame(np.array(pd.read_csv('angiosperm_matrix.csv'))))))
angio_dat.transpose().plot(legend = False, linewidth = 3,alpha = 0.5)
```

```python
angio_gmm1 = fit_GMM(np.array(angio_dat))
angio_vbgmm1 = fit_VBGMM(np.array(angio_dat))
angio_dppgmm1 = fit_DPGMM(np.array(angio_dat))
```

```python
print angio_gmm1[2]
print angio_vbgmm1[2]
print angio_dppgmm1[2]
```

```python
svd_angio = np.linalg.svd(angio_dat)
#gmmcols = np.array(['r', 'b', 'y', 'g'])[list(angio_gmm1[0].predict(angio_dat))]
cols = np.array(['r', 'b', 'y', 'g'])[list(angio_vbgmm1[1].predict(angio_dat))]



fig, axs = plt.subplots(1, 2)

for i in range(angio_dat.shape[0]):
    angio_dat.ix[i, :].transpose().plot(ax = axs[0], legend = False, color = cols[i], alpha = 0.7, linewidth = 2)
    axs[0].set_ylabel('log10 branch length')
    axs[0].set_xlabel('branch')
    axs[1].plot(svd_angio[0][i, 0], svd_angio[0][i, 1], cols[i]+'o', ms = 20, alpha =0.5)
    axs[1].set_ylabel('PC 1')
    axs[1].set_xlabel('PC 2')
pylab.savefig('Angiosperm.pdf')
```

## Poaceae

```python
poac_data = pd.DataFrame(rescale_data(np.array(pd.read_csv('poaceae_matrix.csv'))))
poac_data.transpose().plot(legend = False, linewidth = 2, alpha = 0.5)
```

```python
poac_GMM = fit_GMM(np.array(poac_data))
poac_VBGMM = fit_VBGMM(np.array(poac_data))
poac_DPGMM = fit_DPGMM(np.array(poac_data))
```

```python
print poac_GMM[2]
print poac_VBGMM[2]
print poac_DPGMM[2]
```

```python
svd_poac = np.linalg.svd(poac_data)
cols = np.array(['ro', 'bo', 'yo', 'go', 'r^', 'b^', 'y^', 'g^', 'r*', 'b*', 'y*', 'g*', 'r+', 'b+', 'y+', 'g+'])[list(poac_VBGMM[1].predict(poac_data))]

fig, axs = plt.subplots(1, 2)

for i in range(poac_data.shape[0]):
    poac_data.ix[i, :].transpose().plot(ax = axs[0], legend = False, color = cols[i][0], alpha = 0.7, linewidth = 2)
    axs[1].plot(svd_poac[0][i, 0], svd_poac[0][i, 1], cols[i], ms = 20, alpha =0.5)
    axs[0].set_ylabel('log10 branch length')
    axs[0].set_xlabel('branch')
    axs[1].set_ylabel('PC 1')
    axs[1].set_xlabel('PC 2')
pylab.savefig('Poaceae.pdf')
```

```python
execfile('GMM_trees.py')
# Asteraceae
aster_data = pd.DataFrame(rescale_data(np.array(pd.read_csv('asteraceae_matrix.csv'))))
aster_data.transpose().plot(legend = False, linewidth = 3, alpha = 0.5)
#print aster_data.head()
#print pd.read_csv('asteraceae_matrix.csv')
```

## Eudicots

## Rosids

```python
rosids_data = pd.DataFrame(rescale_data(np.array(pd.read_csv('rosids_matrix.csv'))))
rosids_data.transpose().plot(legend = False, linewidth = 3, alpha = 0.7)
```

```python
print rosids_GMM[2]
print rosids_VGMM[2]
print rosids_DPP[2]
```

```python
rosids_svd = np.linalg.svd(rosids_data)
cols = np.array(['ro', 'bo', 'yo', 'go', 'mo', 'co', 'ko', 'k^'])[rosids_VGMM[1].predict(rosids_data)]

fig, axs = plt.subplots(1, 2)
for i in range(rosids_data.shape[0]):
    rosids_data.ix[i, :].plot(color = cols[i][0], ax = axs[0], linewidth = 3, alpha = 0.7)
    plt.plot(rosids_svd[0][i, 0], rosids_svd[0][i, 1], cols[i], ms = 20, alpha = 0.7)
    axs[0].set_ylabel('log10 branch length')
    axs[0].set_xlabel('branch')
    axs[1].set_ylabel('PC 1')
    axs[1].set_xlabel('PC 2')
pylab.savefig('rosids.pdf')
```
