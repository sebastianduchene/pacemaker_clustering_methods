## Asteraceae

Load data

```python
aster_data = pd.DataFrame(rescale_data(np.array(pd.read_csv('asteraceae_matrix.csv'))))
aster_data.transpose().plot(legend = False, linewidth = 3, alpha = 0.5)
```

**Fig 1.** Plot of branch lengths along the *x* axis using different colours. The *y* axis is the $log_{10}$ branch length and the *x* axis is the different branches. Each line is a gene tree, such that those that are close together along the *x* axis should correspond to the same pacemaker.

Fit the VBGMM model to the angiosperm data

```python
aster_VBGMM = fit_VBGMM(np.array(aster_data))
aster_DPGMM = fit_DPGMM(np.array(aster_data))
```

The output is a list with a set of results. The second item of the list is a table with some summary statistics

```python
print aster_VBGMM[2]
```

The results can be plotted using some items from the lists above. The best fitting model, the VBGMM with a spherical covariance, can be used to plot the resutls.

```python
aster_svd = np.linalg.svd(aster_data)

cols = np.array(['ro', 'yo', 'go', 'bo', 'co', 'mo', 'ko', 'r^', 'y^', 'm^'])[list(aster_VBGMM[1].predict(aster_data))]

fig, axs = plt.subplots(1, 2)
for i in range(aster_data.shape[0]):
    aster_data.ix[i, :].plot(color = cols[i][0], ax = axs[0], linewidth = 2, alpha = 0.7)
    axs[1].plot(aster_svd[0][i, 0], aster_svd[0][i, 1], cols[i], ms = 20, alpha = 0.7)
    
    axs[0].set_ylabel('log10 branch length')
    axs[0].set_xlabel('branch')
    axs[1].set_ylabel('PC 1')
    axs[1].set_xlabel('PC 2')
pylab.savefig('Asteraceae.pdf')
```

**FIg 2.** The first plot is similar to that in Fig 1. In this case, however, the colour corresponds to the cluster assigned. The second plot is the first two components of the singular value decomposition of the branch lengths. Each dot corresponds to a gene tree and the colours represent the different pacemakers.
